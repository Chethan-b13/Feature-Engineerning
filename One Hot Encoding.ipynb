### One hot Encoding for features having many catogories

```python
df = pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
for clmn in df.columns:
    print(f'{clmn} ---- {len(df[clmn].unique())}')
```

#### Applying one hot encoding for only first 10 categories

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
first_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
first_10 = list(first_10)
```

```python
for category in first_10:
    df[category] = np.where(df['X1']==category,1,0)
```

```python
first_10.append('X1')
```
